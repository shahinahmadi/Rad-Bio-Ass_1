# Comparison of stopping power of electron in water, skin and brain

```python
from pathlib import Path
import os
import pandas as pd
import matplotlib.pyplot as plt
path = Path("./Data")
print(path.parent)
DF1 = pd.read_excel("./Data/Electron_Water.xlsx")
DF2=pd.read_excel("./Data/Electron_Brain.xlsx")
DF3=pd.read_excel("./Data/Electron_Skin.xlsx")
plt.scatter(DF1.Energy,DF1.TotalStp,color="none",marker="s", edgecolor="green")
plt.scatter(DF2.Energy,DF2.TotalStp,color="none",marker="^", edgecolor="yellow")
plt.scatter(DF3.Energy,DF3.TotalStp,color="none", edgecolor="blue")
plt.scatter(DF1.Energy,DF2.TotalStp/DF1.TotalStp,color="none",marker="^", edgecolor="green")
plt.scatter(DF1.Energy,DF3.TotalStp/DF1.TotalStp,color="none",marker="s", edgecolor="yellow")
plt.xlabel('Energy (MeV)')
plt.ylabel('Total Stp. Pow. MeV cm2/g')
plt.legend(['Water','Brain','Skin', "Skin/water","Brain/water"])
#plt.ylim(0,2.5)
#plt.xlim(0,20)
plt.xscale("log")
plt.yscale("log")
plt.show()
```
