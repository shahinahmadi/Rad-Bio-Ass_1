# PDD for electron

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_excel("Depth_dose.xlsx", sheet_name="Electron")

depth_e=df1["Depth"]
Dose_e=df1["Dose"]
plt.plot(depth_e, Dose_e, color="Red")
plt.xlabel('Depth(cm)')
plt.ylabel('PDD')
plt.title('PDD of electron')
plt.savefig("plot6.pdf", bbox_inches='tight')
plt.show()
```
